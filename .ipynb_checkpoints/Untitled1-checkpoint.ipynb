```python
import pandas as pd
import numpy as np

df = pd.read_csv("Amazon Sale Report.csv", low_memory=False)

# -------------------------------
# 1. DROP unwanted columns
# -------------------------------
cols_to_drop = ["Unnamed: 22", "fulfilled-by"]
df.drop(columns=cols_to_drop, inplace=True)

# -------------------------------
# 2. HANDLE missing 'Courier Status'
# -------------------------------
df = df[df["Courier Status"].notna()]  # remove rows where courier status is missing

# -------------------------------
# 3. HANDLE missing 'promotion-ids'
# -------------------------------
df["promotion-ids"] = df["promotion-ids"].fillna("No Promotion Used")

# -------------------------------
# 4. HANDLE missing 'currency'
# -------------------------------
df["currency"] = df["currency"].fillna("INR")

# -------------------------------
# 5. HANDLE missing 'Amount'
# Logic: missing Amount appears when EasyShip auto-fills.
# Option 1: Fill with median (recommended for general analysis)
# -------------------------------
df["Amount"] = pd.to_numeric(df["Amount"], errors="coerce")   # ensure numeric
median_amount = df["Amount"].median()
df["Amount"] = df["Amount"].fillna(median_amount)

# -------------------------------
# 6. OPTIONAL: Clean ship-city/state small missing %
# -------------------------------
df["ship-city"] = df["ship-city"].fillna("Unknown")
df["ship-state"] = df["ship-state"].fillna("Unknown")
df["ship-country"] = df["ship-country"].fillna("Unknown")

# -------------------------------
# 7. CHECK cleaned dataset
# -------------------------------
df.isna().sum(), df.head()
```

```python
platform_category_sales = (
    df.groupby(["Sales Channel ", "Category"])["Qty"]
      .sum()
      .reset_index()
      .sort_values(by="Qty", ascending=False)
)

platform_category_sales.head(10)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=platform_category_sales, x="Category", y="Qty", hue="Sales Channel ")
plt.title("Sales Volume per Category per Platform")
plt.xticks(rotation=45)
plt.show()
```
