- The dataset provides insights on the sales in the Amazon platform, including SKU code, design number, Stock, Category, size etc to help optimise the profitablity.
- Courier status have 5.3% of missing values. So we are dropping those columns
- Promotion-ids have ~38% of missing values. Here, we can consider the missing values as no promotions used.I would like to use a code "No promotion used"
- There is ~6% missing values in currency column. There are cases where the customer used promotion to purchase. But there are columns where it is not used. The whole dataset is using INR, so we can replace the Nan values with INR
- There are ~6% of missing values in the "Amount" columnn, which are full filled by "Easyship". We can drop this values for now. But in the future, we should explore why this is happening. 72% of data is missing whether it is shipped by easy ship or not. For the moment, we assume it is shipped by easyship and don't want to drop the rows.
- The dataset contains a column named "Unamed: 22" which have two unique values "Nan" and "False". There is insufficient information about this column, hence we decided to drop this column from the further analysis.

```python
# drop column unnamed:22 

df_amazon_sales.drop(columns=["Unnamed: 22"], inplace= True)
```

```python
df_amazon_sales.columns
```

```python
df_amazon_sales = df_amazon_sales[~df_amazon_sales["Courier Status"].isna()] # negation symbol ~
```

```python
df_amazon_sales.isna().sum()
```

```python
# replaced the missing values of the column with a specified value
df_amazon_sales["promotion-ids"].fillna("No promotions used", inplace=True)
```

```python
df_amazon_sales["promotion-ids"].isna().sum()
```

```python
df_amazon_sales["currency"].fillna("INR", inplace= True)
```

```python
df_amazon_sales = df_amazon_sales[~df_amazon_sales["Amount"].isna()]
```

```python
df_amazon_sales = df_amazon_sales[~df_amazon_sales["ship-country"].isna()]
```

```python
df_amazon_sales.isna().sum()
```

```python
df_amazon_sales["fulfilled-by"].unique()
```

```python
df_amazon_sales["fulfilled-by"].fillna("Easy Ship", inplace= True)
```
